```python
# SST-2 Dataset
# Binary Sentiment Classification (Positive / Negative)

SST_2 = [
    {
        "id": 1,
        "sentence": "This movie was absolutely wonderful and full of heart.",
        "label": 1
    },
    {
        "id": 2,
        "sentence": "The film was boring and far too long.",
        "label": 0
    },
    {
        "id": 3,
        "sentence": "An entertaining story with strong performances.",
        "label": 1
    },
    {
        "id": 4,
        "sentence": "The plot made no sense and the acting was terrible.",
        "label": 0
    },
    {
        "id": 5,
        "sentence": "A beautifully directed movie that keeps you engaged.",
        "label": 1
    },
    {
        "id": 6,
        "sentence": "This was a disappointing experience overall.",
        "label": 0
    }
]
```

```python
# usage

for sample in SST_2:
    sentiment = "Positive" if sample["label"] == 1 else "Negative"
    print(sentiment, "→", sample["sentence"])
```
