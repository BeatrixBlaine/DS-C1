```python
# Grade School Math 8K Dataset
# Author: You
# Description: A dataset of grade-school level math problems

import random

DATASET_TITLE = "grade school math 8k"

topics = ["addition", "subtraction", "multiplication", "division", "word_problem"]
difficulties = ["easy", "medium"]

def generate_problem(idx):
    topic = random.choice(topics)
    difficulty = random.choice(difficulties)

    if topic == "addition":
        a, b = random.randint(1, 50), random.randint(1, 50)
        question = f"What is {a} + {b}?"
        answer = str(a + b)

    elif topic == "subtraction":
        a, b = random.randint(10, 60), random.randint(1, 50)
        question = f"What is {a} - {b}?"
        answer = str(a - b)

    elif topic == "multiplication":
        a, b = random.randint(1, 12), random.randint(1, 12)
        question = f"What is {a} ร {b}?"
        answer = str(a * b)

    elif topic == "division":
        b = random.randint(1, 12)
        a = b * random.randint(1, 12)
        question = f"What is {a} รท {b}?"
        answer = str(a // b)

    else:  # word_problem
        a, b = random.randint(1, 20), random.randint(1, 20)
        question = (
            f"Sarah has {a} apples and buys {b} more. "
            f"How many apples does she have now?"
        )
        answer = str(a + b)

    return {
        "id": idx,
        "question": question,
        "answer": answer,
        "difficulty": difficulty,
        "topic": topic
    }


def build_dataset(size=8000):
    return [generate_problem(i) for i in range(size)]


# Build the dataset
grade_school_math_8k = build_dataset()

# Example sample
print(DATASET_TITLE)
print(grade_school_math_8k[0])
```
