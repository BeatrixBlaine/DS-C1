```python
# SST-5 Dataset
# Fine-grained Sentiment Classification (5 classes)

SST_5 = [
    {
        "id": 1,
        "sentence": "This movie is an absolute masterpiece.",
        "label": 4
    },
    {
        "id": 2,
        "sentence": "A wonderful and emotionally rich experience.",
        "label": 4
    },
    {
        "id": 3,
        "sentence": "The film was enjoyable but not particularly memorable.",
        "label": 3
    },
    {
        "id": 4,
        "sentence": "The movie was okay, nothing special.",
        "label": 2
    },
    {
        "id": 5,
        "sentence": "The plot was weak and the acting average.",
        "label": 1
    },
    {
        "id": 6,
        "sentence": "This was a painful and frustrating movie to watch.",
        "label": 0
    }
]
```

```python
# labels

SST_5_LABELS = {
    0: "very negative",
    1: "negative",
    2: "neutral",
    3: "positive",
    4: "very positive"
}
```

```python
# usage
for sample in SST_5:
    print(SST_5_LABELS[sample["label"]], "→", sample["sentence"])
```
