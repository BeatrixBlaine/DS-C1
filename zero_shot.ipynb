```python
from transformers import pipeline
from pprint import pprint

# Create the zero-shot classification pipeline
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
```

```python
# sample text
sequence = "I absolutely love the new smartphone design - it's sleek and super intuitive!"

# define candidate labels for sentiment analysis
candidate_labels = ["positive", "negative", "neutral"]

# result for zero shot classification on the sequence with defined candidate labels
result = classifier(sequence, candidate_labels)

# output
pprint(result)
```

```python
# find the index of the highest scores
sentiment = result['labels'][result['scores'].index(max(result['scores']))]

# print the best result
print(f"\nExtracted Sentiment -> {sentiment}")
```
