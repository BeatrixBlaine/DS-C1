```python
# Random Example Generator

import random

def random_math_example():
    ops = ["+", "-", "×"]
    a = random.randint(1, 10)
    b = random.randint(1, 10)
    op = random.choice(ops)

    if op == "+":
        answer = a + b
    elif op == "-":
        answer = a - b
    else:
        answer = a * b

    question = f"What is {a} {op} {b}?"
    return question, str(answer)
```

```python
# Few Shot Prompt

def build_random_few_shot_prompt(k, query):
    prompt = "You are a grade school math solver.\n\n"

    for _ in range(k):
        q, a = random_math_example()
        prompt += f"Q: {q}\nA: {a}\n\n"

    prompt += f"Q: {query}\nA:"
    return prompt
```

```python
# Usage

prompt = build_random_few_shot_prompt(
    k=4,
    query="What is 8 + 9?"
)

print(prompt)
```
